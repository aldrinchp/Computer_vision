### **Step 2: Import Required Libraries**

Import the libraries you will use:

```python
import cv2
from matplotlib import pyplot as plt
```

### **Step 3: Load the Image**

Use OpenCV's `imread` function to load your image.

Replace the path with the location of your image file.

```python
# Load the image (change the path to your image file)
image_bgr = cv2.imread('/media/aldrinchp/KAli-HDD/Subjects/NOVENO/Computer vision/Computer_Vision/UC.04 Object Detection/Object Detection/cat.jpg')
```

```python
# Print the shape of the image (height, width, channels)
print('Image shape:', image_bgr.shape)
```

### **Step 4: Show the image**

```python
# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.show()
```

### **Step 5: Create the Bounding Box over the Image**

To draw a bounding box, use OpenCV's `rectangle` function.  

You need to specify the coordinates of the **top-left** point $(x1, y1)$ and the **bottom-right** point $(x2, y2)$ of the bounding box.

The coordinates should be within the image dimensions

```python
x1, y1 = 250, 200  # Top-left corner
x2, y2 = 700, 900  # Bottom-right corner

# Draw the bounding box on the input image using cv2.rectangle
# Arguments: image, (x1, y1), (x2, y2), bounding box color in BGR, thickness
image_bgr_box = cv2.rectangle(image_bgr, (x1, y1), (x2, y2), (0, 255, 0), 15)  # Green box, thickness 15
```

### **Step 6: Display the Image and Bounding Box**

Finally, use matplotlib to display the image with the bounding box.

```python
# Convert the image from BGR (OpenCV format) to RGB (matplotlib format)
image_rgb_box = cv2.cvtColor(image_bgr_box, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb_box)
plt.show()
```

You should now see your image with a bounding box drawn around the object.  

You can adjust the bounding box coordinates as needed to fit the object you want to highlight.

### **Individual Activity:**

1. Load an image containing **at least three different objects from different classes** using `cv2.imread`.

```python
image_minions = cv2.imread('/media/aldrinchp/KAli-HDD/Subjects/NOVENO/Computer vision/Computer_Vision/UC.04 Object Detection/Minions.jpeg')
print('Image shape:', image_minions.shape)
```

```python
image_minions_rgb = cv2.cvtColor(image_minions, cv2.COLOR_BGR2RGB)
plt.imshow(image_minions_rgb)
plt.show()
```

2. Draw **different colored bounding boxes** around each object (at least one of the bounding boxes should be yellow). Use a **line thickness** of 5 for both bounding boxes.

```python
# Copia para no modificar la imagen original
image_minions_boxes = image_minions.copy()

# Coordenadas (x1, y1, x2, y2) para tres objetos
bbox_minion_left = (20, 40, 80, 140)
bbox_minion_right = (120, 50, 170, 140)
bbox_minion_last = (220, 20, 270, 140)

# Dibujar bounding boxes con colores diferentes (BGR) y grosor 5
cv2.rectangle(
    image_minions_boxes,
    (bbox_minion_left[0], bbox_minion_left[1]),
    (bbox_minion_left[2], bbox_minion_left[3]),
    (0, 255, 0),
    5
)

cv2.rectangle(
    image_minions_boxes,
    (bbox_minion_right[0], bbox_minion_right[1]),
    (bbox_minion_right[2], bbox_minion_right[3]),
    (255, 0, 0),
    5
)

cv2.rectangle(
    image_minions_boxes,
    (bbox_minion_last[0], bbox_minion_last[1]),
    (bbox_minion_last[2], bbox_minion_last[3]),
    (0, 255, 255),
    5
)
```

3. Add **text labels** above each bounding box (e.g., “Cat”, “Ball”, etc.) using `cv2.putText`.

```python
# Crear una copia para agregar etiquetas
image_minions_labeled = image_minions_boxes.copy()

font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.2
font_scale_small = 0.7
thickness = 3

# Etiquetas encima de cada bounding box
cv2.putText(
    image_minions_labeled,
    'Minion 1',
    (bbox_minion_left[0], bbox_minion_left[1] - 15),
    font,
    font_scale,
    (0, 255, 0),
    thickness,
    cv2.LINE_AA
)

cv2.putText(
    image_minions_labeled,
    'Minion 2',
    (bbox_minion_right[0], bbox_minion_right[1] - 15),
    font,
    font_scale,
    (255, 0, 0),
    thickness,
    cv2.LINE_AA
)

cv2.putText(
    image_minions_labeled,
    'Minion 3',
    (bbox_minion_last[0], bbox_minion_last[1] - 10),
    font,
    font_scale_small,
    (0, 255, 255),
    thickness,
    cv2.LINE_AA
)
```

4. Display the **resulting image**.

```python
# Convertir de BGR (OpenCV) a RGB (matplotlib)
image_minions_labeled_rgb = cv2.cvtColor(image_minions_labeled, cv2.COLOR_BGR2RGB)

# Mostrar resultado final
plt.figure(figsize=(8, 8))
plt.imshow(image_minions_labeled_rgb)
plt.axis('off')
plt.title('Minions con Bounding Boxes y Etiquetas')
plt.show()
```

5. Suggest one method (written, no code) to **detect objects automatically** in an image (no manual coordinates). Name the model or technique you would use and briefly explain how it produces bounding boxes.

A good method is **YOLO (You Only Look Once)** for automatic object detection.

YOLO processes the entire image in one pass with a convolutional neural network and directly predicts multiple **bounding boxes**, object confidence scores, and class probabilities. Then, it applies **Non-Maximum Suppression (NMS)** to keep the best box per object and remove overlapping duplicates.

This allows fast and accurate detection without manually choosing coordinates.
