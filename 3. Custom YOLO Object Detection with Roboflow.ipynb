### **Step 9: Test Your Custom Model on Video**

```python
from pathlib import Path
import cv2
from ultralytics import YOLO

# 1) Ruta fija al modelo entrenado
model_path = Path("runs/detect/custom_yolo_model/weights/best.pt")
if not model_path.exists():
    raise FileNotFoundError(f"No se encontró el modelo en: {model_path}")

# 2) Cargar modelo
model = YOLO(str(model_path))
classNames = model.names

# 3) Abrir cámara (intenta 0, luego 1)
captura = None
for cam_idx in [0, 1]:
    test_cap = cv2.VideoCapture(cam_idx)
    if test_cap.isOpened():
        captura = test_cap
        print(f"Cámara abierta en índice {cam_idx}")
        break
    test_cap.release()

if captura is None:
    raise RuntimeError("No se pudo abrir ninguna cámara (índices 0 o 1).")

# 4) Parámetros (parecido al notebook 2, pero con el modelo custom)
imgsz = 416
conf = 0.30
print("Detección iniciada. Presiona 'q' para salir.")

while True:
    success, img = captura.read()
    if not success:
        print("No se pudo leer un frame de la cámara.")
        break

    results = model(img, stream=True, imgsz=imgsz, conf=conf, verbose=False)

    for r in results:
        boxes = r.boxes
        for box in boxes:
            cls = int(box.cls[0])
            class_name = classNames[cls]

            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 2)
            label = f"{class_name} {float(box.conf[0]):.2f}"
            cv2.putText(
                img,
                label,
                (x1, max(y1 - 10, 20)),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.7,
                (255, 0, 0),
                2
            )

    cv2.imshow('Webcam', img)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

captura.release()
cv2.destroyAllWindows()
```
